# Removing outlier 

```python
def cap_outliers_iqr(df, columns):
    for column in columns:
        Q1 = df[column].quantile(0.25)
        Q3 = df[column].quantile(0.75)
        IQR = Q3 - Q1
        lower = Q1 - 1.5 * IQR
        upper = Q3 + 1.5 * IQR

        # Cap values for each column separately
        df[column] = df[column].clip(lower, upper)
        print(f"Capped outliers in {column}")
    return df

# Apply only to km and Mileage
df = cap_outliers_iqr(df, ['km', 'Mileage'])

print("Outlier capping done âœ…")
```
